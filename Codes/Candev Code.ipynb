```python
#importing the data
import pandas as pd
import nltk
import numpy as np
from pandas import DataFrame
```

```python
#reading the dataset
dataset = pd.read_csv("GcCampus metadata.csv", encoding = "ISO-8859-1")
dataset.head(10)
```

```python
keyword = ['Cyber','IT','Risk','Human','Ethics']

title_list = dataset['title'].tolist()
frequence = 0
matrix_list = [0]*len(title_list)

for key in keyword:
    for i in range (0, len(title_list)-1):
        x = str(title_list[i])
        sub_index = x.find(key)
        
        if(sub_index != -1):
            #print (x,"--->",key)
            matrix_list[i] += 1

description_list = dataset['Description'].tolist()
for key in keyword:
    for i in range (0, len(description_list)-1):
        x = str(description_list[i])
        sub_index = x.find(key)
        
        if(sub_index != -1):
            #print (title_list[i],"--->",key)
            #print(key)
            matrix_list[i] += 1
            
sort_list = sorted(zip(matrix_list, title_list), key=lambda x: x[0], reverse=True)

df = DataFrame(sort_list)
df.columns = ['Frequency', 'Title']
df.head(10)
```

```python
# Program to measure similarity between  
# two sentences using cosine similarity. 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
  
# X = input("Enter first string: ").lower() 
# Y = input("Enter second string: ").lower() 
for i in range (0,len(title_list)-1):
    for j in range (0,len(title_list)-1):
        if(matrix_list[i] != -1):
            X = str(title_list[i])
        if(matrix_list[j] != -1):
            Y = str(title_list[j])
  
# tokenization 
        X_list = word_tokenize(X)  
        Y_list = word_tokenize(Y) 
  
# sw contains the list of stopwords 
        sw = stopwords.words('english')  
        l1 =[];l2 =[] 
  
# remove stop words from string 
        X_set = {w for w in X_list if not w in sw}  
        Y_set = {w for w in Y_list if not w in sw} 
  
# form a set containing keywords of both strings  
        rvector = X_set.union(Y_set)  
        for w in rvector: 
            if w in X_set: l1.append(1) # create a vector 
            else: l1.append(0) 
            if w in Y_set: l2.append(1) 
            else: l2.append(0) 
        c = 0
  
# cosine formula  
        for i in range(len(rvector)):
            c+= l1[i]*l2[i] 
        cosine = round((c / float((sum(l1)*sum(l2))**0.5))*100)
        if (cosine > 0.5 and cosine != 1):
            print("similarity rate between",title_list[i],"and", title_list[j], "is: ", cosine, "%") 
```
